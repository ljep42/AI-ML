```python
class_names = {0: 'Not Fraud', 1: 'Fraud'}
print(df.Class.value_counts().rename(index=class_names))

print(df.iloc[:,2])
```

```python
# define figure
fig, axes = plt.subplots(nrows=5, ncols=6)

# loop through subplot and plot dataframe
k = 1
for i in range(5):
    for j in range(6):
        if k < 30:
            axes[i][j].plot(df.iloc[:,k])
            k += 1
        else:
            next

# show figure
plt.show()
```

```python
# define variables for features and target
feature_names = df.iloc[:1,:30].columns
target = df.iloc[:1,30:].columns
print(feature_names)
print(target)

data_features = df[feature_names]
data_target = df[target]
```

```python
from sklearn.model_selection import train_test_split

# split into test train, default to 75/25
X_train, X_test, y_train, y_test = train_test_split(data_features, data_target)

# check
print("length of x_train: {x_train}".format(x_train = len(X_train)))
print("length of x_test: {x_test}".format(x_test = len(X_test)))
print("length of y_train: {y_trn}".format(y_trn = len(y_train)))
print("length of y_test: {y_trn}".format(y_trn = len(y_test)))

print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report

# define model, optimizer is liblinear
model = LogisticRegression(solver='liblinear')
# fit to train data
model.fit(X_train, y_train.values.ravel())
# predict test data
preds = model.predict(X_test)
```

```python
# get confusion matrix from true values and predictions
matrix = confusion_matrix(y_test, preds)
print(matrix)
```

```python
# create heatmap
sns.heatmap(matrix, annot=True, cmap="Blues", fmt='g')
# add graph features
plt.title("Confusion Matrix"), plt.tight_layout(), plt.ylabel("True"), plt.xlabel("Predicted")
plt.show()
```

```python
# get report with recall, precision, F1-score
print(classification_report(y_test, preds))
```
